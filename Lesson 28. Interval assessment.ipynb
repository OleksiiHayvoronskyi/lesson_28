### Завдання 1
 
- [x] Із генеральної сукупності, розподіленої за нормальним законом, отримано вибірку обсягом $ n = 20 $.
   * Знайти з надійністю $ \gamma = 0,95$ інтервальну оцінку для середнього квадратичного відхилення $ σ $ генеральної сукупності.


**Отримана вибірка:** $ \\ $

|         |   |   |   |    |    |
| :-:     |:-:|:-:|:-:|:-: |:-: |
| $ x_i $ | 1 | 2 | 4 | 6  | 7  |
| $ n_i $ | 2 | 3 | 10| 3  | 2  |

**Розв’язання**: $ \\ $

*Знаходимо ВИБІРКОВЕ ЗНАЧЕННЯ*: $ \\ $
$ x^- _в = \frac{(1 \mathop ⋅ 2 \mathop + 2 \mathop ⋅ 3 \mathop + 4 \mathop ⋅ 10 \mathop + 6 \mathop ⋅ 3 \mathop + 7 \mathop ⋅ 2)} {20} = 4 $. $ \\ $

*Знаходимо ВИПРАВЛЕНУ ДИСПЕРСІЮ за формулою*: $ \\ $
$ s_2 = \frac{1}{n-1} \sum^k _{i=1} (x_i - x^- _в)^2 \mathop{n_i}$ :  $ \\ $ 

$ s_2 = \frac{((1 \mathop - 4)^2 \mathop ⋅ 2 \mathop + (2\mathop - 4)^2 \mathop ⋅ 3 \mathop + (4 \mathop - 4)^2 \mathop ⋅ 10 \mathop + (6 \mathop - 4)^2 \mathop ⋅ 3 \mathop + (7 \mathop - 4)^2 \mathop ⋅ 2)} {19} = 3,2 $. $ \\ $
$ s^2 = 3,2 $.

*Знаходимо ВИПРАВЛЕНЕ СЕРЕДНЄ КВАДРАТИЧНЕ ВІДХИЛЕННЯ за формулою*: $ \\ $
$ s = \sqrt{s^2} = \sqrt{3,2} \approx 1,79$. $ \\ $

*Із Таблиці значень $ q = q\mathop(y,\mathop n)$ знаходимо $ q $ :* $ \\ $
$ q = 0,37 $

*Знаходимо ДОВІРЧИЙ ІНТЕРВАЛ за формулою*: $ \\ $
$ s \mathop (1 - q) < \sigma < s \mathop (1 + q) $: $ \\ $
$ s \mathop (1 - q) = 1,79 ⋅ \mathop (1 - 0,37) \approx 1,13 $; $ \\ $
$ s \mathop (1 + q) = 1,79 ⋅ \mathop (1 + 0,37) \approx 2,45 $. $ \\ $


**Відповідь**: $ \\ $
*Інтервальна оцінка для середнього квадратичного відхилення $ σ $ генеральної сукупності* $ = 1,13 < \sigma < 2,45 $.


### Завдання 2

- [x] 
    * Знайти з надійністю $ \gamma = 0,95$ інтервальну оцінку для ймовірності настання події $ A $ в кожному із $ n = 100 $ незалежних повторних випробувань, якщо подія відбулася $ m = 40 $ разів. $ \\ $


**Розв’язання**: $\\ $

*З Таблиці значень функції Лапласа знаходимо АРГУМЕНТ ФУНКЦІЇ ЛАПЛАСА ($ t $):* $\\ $
$ Ф(t) = \frac{\gamma} {2} = \frac{0,95} {2} = 0,475 $. $ \\ $
$ t = 1,96 $. $\\ $

*Обчислюємо МЕЖІ ДОВІРЧОГО ІНТЕРВАЛУ* $ p_1 < p < p_2 $ за формулою: $\\ $
$ p_1 = \frac{1} {t^2 \mathop + n} \mathop (m \mathop + \frac{t^2} {2} \mathop - t \sqrt {\frac{m \mathop(n \mathop - m)} {n} \mathop + \frac {t^2} {4}}) $ : $ \\ $

$ p_1 = \frac{1} {1,96^2 \mathop + 100} \mathop (40 \mathop + \frac{1,96^2} {2} \mathop - 1,96 \sqrt {\frac{40 \mathop(100 \mathop - 40)} {100} \mathop + \frac {1,96^2} {4}}) = 0,31 $. $ \\ $
$ p_1 = 0,31 $. $ \\ $

$ p_2 = \frac{1} {t^2 \mathop + n} \mathop (m \mathop + \frac{t^2} {2} \mathop + t \sqrt {\frac{m \mathop(n \mathop - m)} {n} \mathop + \frac {t^2} {4}}) $ : $ \\ $

$ p_2 = \frac{1} {1,96^2 \mathop + 100} \mathop (40 \mathop + \frac{1,96^2} {2} \mathop + 1,96 \sqrt {\frac{40 \mathop(100 \mathop - 40)} {100} \mathop + \frac {1,96^2} {4}}) = 0,49 $. $ \\ $
$ p_2 = 0,49 $. $ \\ $

**Відповідь**: $ \\ $
*Інтервальною оцінкою для ймовірності настання події $ A $ є довірчий інтервал:* $ 0,31 < p < 0,49 $.

**Розв’язання**: $ \\ $

*За Таблицею значень функції Лапласа знаходимо АРГУМЕНТ ФУНКЦІЇ ЛАПЛАСА ($ t $)*: $ \\ $

$ t = 2,58 $. $ \\ $

*За формулою $ n = \frac{t^2 \mathop ⋅ \sigma^2} {\delta^2}$*: $ \\ $
$ n = \frac{(2,58)^2 \mathop ⋅ 3^2} {0,6^2} \approx 166 $. $ \\ $

**Відповідь**: $ \\ $
*Обсяг вибірки дорівнює $ 166 $*.
